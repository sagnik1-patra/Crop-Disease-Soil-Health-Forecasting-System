```python
# ===============================================================
# ðŸŒ± AGRISENTINEL â€” Result Display & Visualization
# ===============================================================

import os
import json
import joblib
import yaml
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model

# ---------------------------------------------------------------
# ðŸ§© Paths
# ---------------------------------------------------------------
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Crop Disease & Soil Health Forecasting System"
DATA_PATH = os.path.join(BASE_DIR, "archive", "dataset.csv")

MODEL_PATH = os.path.join(BASE_DIR, "agrisentinel_model.h5")
SCALER_PATH = os.path.join(BASE_DIR, "soil_scaler.pkl")
REPORT_PATH = os.path.join(BASE_DIR, "agrisentinel_report.json")
CONFIG_PATH = os.path.join(BASE_DIR, "config.yaml")

# ---------------------------------------------------------------
# ðŸ§  Load Saved Artifacts
# ---------------------------------------------------------------
print("[INFO] Loading model, scaler, and reports...")

model = load_model(MODEL_PATH)
scaler_x = joblib.load(SCALER_PATH)

with open(REPORT_PATH, "r") as f:
    report = json.load(f)

with open(CONFIG_PATH, "r") as f:
    config = yaml.safe_load(f)

print("[INFO] âœ… Model & configs loaded successfully!")

# ---------------------------------------------------------------
# ðŸ“¥ Load Dataset
# ---------------------------------------------------------------
df = pd.read_csv(DATA_PATH)
feature_cols = [c for c in df.columns if c.lower() != 'label']
target_col = 'label'

X = df[feature_cols].values
y = df[target_col].values

# Recreate label encoder
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

# ---------------------------------------------------------------
# ðŸ”® Predict Using Model
# ---------------------------------------------------------------
X_scaled = scaler_x.transform(X)
y_pred = np.argmax(model.predict(X_scaled), axis=1)

# Convert back to label names if applicable
if hasattr(encoder, "inverse_transform"):
    actual_labels = encoder.inverse_transform(y_encoded)
    predicted_labels = encoder.inverse_transform(y_pred)
else:
    actual_labels = y_encoded
    predicted_labels = y_pred

# ---------------------------------------------------------------
# ðŸ“Š Print Results Summary
# ---------------------------------------------------------------
print("\n=================== AGRISENTINEL RESULTS ===================")
print("ðŸ”¹ Hybrid Model: PSO + AIS Optimized MLP")
print("ðŸ”¹ Configuration:", config)
print("-------------------------------------------------------------")
for k, v in report.items():
    print(f"{k:15}: {v}")
print("-------------------------------------------------------------")
print(f"âœ… Total Samples Evaluated: {len(y)}")
print(f"âœ… Unique Classes Detected: {len(np.unique(y_encoded))}")
print("=============================================================\n")

# ---------------------------------------------------------------
# ðŸ§© Show Sample Predictions
# ---------------------------------------------------------------
sample_df = pd.DataFrame({
    "Actual": actual_labels[:15],
    "Predicted": predicted_labels[:15]
})
print("ðŸ“‹ Sample Predictions (First 15):")
print(sample_df.to_string(index=False))

# ---------------------------------------------------------------
# ðŸ–¼ï¸ Display Saved Graphs
# ---------------------------------------------------------------
import matplotlib.image as mpimg

graphs = [
    "agrisentinel_accuracy_graph.png",
    "agrisentinel_heatmap.png",
    "agrisentinel_prediction_graph.png",
    "agrisentinel_result_graph.png",
    "agrisentinel_comparison_graph.png"
]

for g in graphs:
    graph_path = os.path.join(BASE_DIR, g)
    if os.path.exists(graph_path):
        img = mpimg.imread(graph_path)
        plt.figure(figsize=(7,5))
        plt.imshow(img)
        plt.axis('off')
        plt.title(g.replace("_", " ").replace(".png", "").title(), fontsize=12, pad=10)
        plt.show()
    else:
        print(f"[WARN] Graph not found: {g}")

# ---------------------------------------------------------------
# ðŸ“ˆ Save Combined Comparison CSV
# ---------------------------------------------------------------
comparison_path = os.path.join(BASE_DIR, "agrisentinel_results.csv")
pd.DataFrame({
    "Actual": actual_labels,
    "Predicted": predicted_labels
}).to_csv(comparison_path, index=False)

print(f"[INFO] âœ… Comparison file saved at: {comparison_path}")
print(f"[INFO] âœ… All visualizations displayed successfully!")
```
