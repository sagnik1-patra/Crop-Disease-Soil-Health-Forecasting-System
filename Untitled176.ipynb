```python
import os
import json
import yaml
import joblib
import random
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical

warnings.filterwarnings("ignore")

# ---------------------------------------------------------------
# ðŸ§© Paths
# ---------------------------------------------------------------
DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Crop Disease & Soil Health Forecasting System\archive\dataset.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Crop Disease & Soil Health Forecasting System"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ---------------------------------------------------------------
# ðŸ“¥ Load Dataset
# ---------------------------------------------------------------
print("[INFO] Loading dataset...")
df = pd.read_csv(DATA_PATH)
print("[INFO] Shape:", df.shape)
print("[INFO] Columns:", df.columns.tolist())

# ---------------------------------------------------------------
# ðŸ” Check available features
# ---------------------------------------------------------------
# Use only numeric columns except label
feature_cols = [c for c in df.columns if c.lower() != 'label']
target_col = 'label'

X = df[feature_cols].values
y = df[target_col].values

# ---------------------------------------------------------------
# ðŸ”¤ Encode Label
# ---------------------------------------------------------------
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_categorical = to_categorical(y_encoded)

# ---------------------------------------------------------------
# ðŸ”¢ Scale Features
# ---------------------------------------------------------------
scaler_x = MinMaxScaler()
X_scaled = scaler_x.fit_transform(X)

# ---------------------------------------------------------------
# âœ‚ï¸ Split Data
# ---------------------------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_categorical, test_size=0.2, random_state=42)

# ---------------------------------------------------------------
# ðŸ§  Base MLP Model
# ---------------------------------------------------------------
def build_model(input_dim, num_classes, hidden_neurons=64, lr=0.001):
    model = Sequential([
        Dense(hidden_neurons, activation='relu', input_dim=input_dim),
        Dropout(0.3),
        Dense(hidden_neurons//2, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer=Adam(learning_rate=lr),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# ---------------------------------------------------------------
# ðŸ§¬ Hybrid PSOâ€“AIS Optimization
# ---------------------------------------------------------------
def hybrid_pso_ais(X_train, y_train, X_val, y_val, n_particles=5, n_iter=5):
    best_params = None
    best_acc = 0

    for i in range(n_iter):
        print(f"[PSO] Iteration {i+1}/{n_iter}")
        particles = []

        for p in range(n_particles):
            hidden = random.randint(32, 256)
            lr = random.uniform(0.0005, 0.005)

            model = build_model(X_train.shape[1], y_train.shape[1], hidden, lr)
            model.fit(X_train, y_train, epochs=10, batch_size=16, verbose=0)
            acc = model.evaluate(X_val, y_val, verbose=0)[1]
            particles.append((hidden, lr, acc))
            print(f"   Particle {p+1}: Hidden={hidden}, LR={lr:.5f}, ACC={acc:.4f}")

        # Select best particle
        particles.sort(key=lambda x: x[2], reverse=True)
        best_particle = particles[0]

        # AIS mutation around best particle
        mutated_hidden = int(best_particle[0] * random.uniform(0.8, 1.2))
        mutated_lr = best_particle[1] * random.uniform(0.8, 1.2)

        print(f"[AIS] Mutated Hidden={mutated_hidden}, LR={mutated_lr:.5f}")

        model = build_model(X_train.shape[1], y_train.shape[1], mutated_hidden, mutated_lr)
        model.fit(X_train, y_train, epochs=10, batch_size=16, verbose=0)
        acc = model.evaluate(X_val, y_val, verbose=0)[1]

        if acc > best_acc:
            best_acc = acc
            best_params = {"hidden": mutated_hidden, "lr": mutated_lr}

        print(f"[RESULT] Iter {i+1} | Best Acc={best_acc:.4f}")

    return best_params

# ---------------------------------------------------------------
# ðŸš€ Run Optimization
# ---------------------------------------------------------------
params = hybrid_pso_ais(X_train, y_train, X_test, y_test)
print("[INFO] Best Params:", params)

# ---------------------------------------------------------------
# ðŸ§  Final Model Training
# ---------------------------------------------------------------
final_model = build_model(X_train.shape[1], y_train.shape[1],
                          hidden_neurons=params["hidden"],
                          lr=params["lr"])

history = final_model.fit(X_train, y_train, validation_data=(X_test, y_test),
                          epochs=30, batch_size=16, verbose=1)

# ---------------------------------------------------------------
# ðŸ“Š Evaluate Model
# ---------------------------------------------------------------
y_pred = np.argmax(final_model.predict(X_test), axis=1)
y_true = np.argmax(y_test, axis=1)

accuracy = accuracy_score(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
r2 = r2_score(y_true, y_pred)

summary_report = {
    "Hybrid_Model": "PSO + AIS Optimized MLP",
    "Accuracy": round(accuracy * 100, 2),
    "RMSE": round(rmse, 3),
    "R2_Score": round(r2, 3),
    "Best_Params": params
}

# ---------------------------------------------------------------
# ðŸ’¾ Save Artifacts
# ---------------------------------------------------------------
model_path = os.path.join(OUTPUT_DIR, "agrisentinel_model.h5")
scaler_path = os.path.join(OUTPUT_DIR, "soil_scaler.pkl")
config_path = os.path.join(OUTPUT_DIR, "config.yaml")
report_path = os.path.join(OUTPUT_DIR, "agrisentinel_report.json")

final_model.save(model_path)
joblib.dump(scaler_x, scaler_path)

with open(config_path, "w") as f:
    yaml.dump(params, f)

with open(report_path, "w") as f:
    json.dump(summary_report, f, indent=4)

print("[INFO] âœ… Model & reports saved successfully!")

# ---------------------------------------------------------------
# ðŸ“ˆ Graphs
# ---------------------------------------------------------------
plt.figure(figsize=(6,4))
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title("AgriSentinel Accuracy vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.savefig(os.path.join(OUTPUT_DIR, "agrisentinel_accuracy_graph.png"))
plt.close()

corr = pd.DataFrame(X_scaled, columns=feature_cols).corr()
plt.figure(figsize=(7,6))
sns.heatmap(corr, annot=True, cmap="YlGnBu")
plt.title("Soil Feature Correlation Heatmap")
plt.savefig(os.path.join(OUTPUT_DIR, "agrisentinel_heatmap.png"))
plt.close()

plt.figure(figsize=(6,4))
plt.plot(y_true[:100], label='Actual')
plt.plot(y_pred[:100], label='Predicted')
plt.title("Prediction Comparison")
plt.xlabel("Sample Index")
plt.ylabel("Class")
plt.legend()
plt.savefig(os.path.join(OUTPUT_DIR, "agrisentinel_prediction_graph.png"))
plt.close()

print("[INFO] ðŸ“Š All graphs generated successfully!")
print("[INFO] Files saved in:", OUTPUT_DIR)
```
