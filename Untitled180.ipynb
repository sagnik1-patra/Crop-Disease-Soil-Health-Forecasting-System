```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from tensorflow.keras.models import load_model
from sklearn.preprocessing import LabelEncoder

# ---------------------------------------------------------------
# ðŸ“‚ Paths
# ---------------------------------------------------------------
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Crop Disease & Soil Health Forecasting System"
DATA_PATH = os.path.join(BASE_DIR, "archive", "dataset.csv")

MODEL_PATH = os.path.join(BASE_DIR, "hybrid_agrisentinel_model.h5")
SCALER_PATH = os.path.join(BASE_DIR, "hybrid_soil_scaler.pkl")
REPORT_PATH = os.path.join(BASE_DIR, "hybrid_agrisentinel_report.json")
CONFIG_PATH = os.path.join(BASE_DIR, "hybrid_config.yaml")

# ---------------------------------------------------------------
# ðŸ§© Load Saved Components
# ---------------------------------------------------------------
print("[INFO] Loading hybrid model, scaler, and configuration...")

model = load_model(MODEL_PATH)
scaler_x = joblib.load(SCALER_PATH)

with open(REPORT_PATH, "r") as f:
    report = json.load(f)

with open(CONFIG_PATH, "r") as f:
    config = yaml.safe_load(f)

print("[INFO] âœ… Model & Config Loaded Successfully!")

# ---------------------------------------------------------------
# ðŸ“¥ Load Dataset for Prediction
# ---------------------------------------------------------------
df = pd.read_csv(DATA_PATH)
feature_cols = [c for c in df.columns if c.lower() != 'label']
target_col = 'label'

X = df[feature_cols].values
y = df[target_col].values

encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

# ---------------------------------------------------------------
# ðŸ”® Predict
# ---------------------------------------------------------------
X_scaled = scaler_x.transform(X)
y_pred = np.argmax(model.predict(X_scaled), axis=1)

# Decode back to labels if possible
try:
    actual_labels = encoder.inverse_transform(y_encoded)
    predicted_labels = encoder.inverse_transform(y_pred)
except:
    actual_labels = y_encoded
    predicted_labels = y_pred

# ---------------------------------------------------------------
# ðŸ“Š Display Hybrid Results Summary
# ---------------------------------------------------------------
print("\n==================== HYBRID AGRISENTINEL RESULTS ====================")
print("Hybrid Model  : AIS + GA Optimized MLP")
print("Configuration :", config)
print("---------------------------------------------------------------------")
for k, v in report.items():
    print(f"{k:15}: {v}")
print("---------------------------------------------------------------------")
print(f"âœ… Total Samples Evaluated : {len(y)}")
print(f"âœ… Unique Classes Detected : {len(np.unique(y_encoded))}")
print("=====================================================================\n")

# ---------------------------------------------------------------
# ðŸ“‹ Sample Predictions
# ---------------------------------------------------------------
sample_df = pd.DataFrame({
    "Actual": actual_labels[:15],
    "Predicted": predicted_labels[:15]
})
print("ðŸ“‹ Sample Predictions (First 15 Rows):")
print(sample_df.to_string(index=False))
print()

# ---------------------------------------------------------------
# ðŸ–¼ï¸ Show All Hybrid Graphs
# ---------------------------------------------------------------
graph_files = [
    "hybrid_agrisentinel_accuracy_graph.png",
    "hybrid_agrisentinel_heatmap.png",
    "hybrid_agrisentinel_prediction_graph.png",
    "hybrid_agrisentinel_result_graph.png",
    "hybrid_agrisentinel_comparison_graph.png"
]

print("[INFO] Displaying all hybrid result graphs...")
for g in graph_files:
    graph_path = os.path.join(BASE_DIR, g)
    if os.path.exists(graph_path):
        img = mpimg.imread(graph_path)
        plt.figure(figsize=(7,5))
        plt.imshow(img)
        plt.axis("off")
        plt.title(g.replace("_", " ").replace(".png", "").title(), fontsize=11, pad=10)
        plt.show()
    else:
        print(f"[WARN] Graph not found: {g}")

# ---------------------------------------------------------------
# ðŸ’¾ Save Combined Results CSV
# ---------------------------------------------------------------
results_path = os.path.join(BASE_DIR, "hybrid_agrisentinel_results.csv")
pd.DataFrame({
    "Actual": actual_labels,
    "Predicted": predicted_labels
}).to_csv(results_path, index=False)

print(f"[INFO] âœ… Hybrid results CSV saved at: {results_path}")
print(f"[INFO] âœ… All hybrid results and graphs displayed successfully!")
```
